## Setup

### Imports

```python
import touchsim as ts # import touchsim package
from touchsim.plotting import plot # import in-built plotting function
import numpy as np
import holoviews as hv # import holoviews package for plots and set some parameters

hv.notebook_extension()
%output holomap='scrubber'

%matplotlib widget

import warnings
warnings.simplefilter("ignore")
```

### Afferent Population

```python
%%output size=150
a = ts.affpop_hand()

plot(a)
```

### Texture

```python
disp_map = ts.texture.DisplacementMap(filename="./textures/tile.jpg", size=(500, 500), max_height=1)

%matplotlib inline
disp_map.visualize()
```

### Path

```python
num_samples = 600

x = np.full(num_samples, disp_map.shape[0] / 4)
y = np.linspace(disp_map.shape[1] / 4, disp_map.shape[1] * 3 / 4, num_samples)
path = np.column_stack((x, y))

disp_map.visualize(path)
```

### Stimulus

```python
loc = np.zeros((1,2))
trace = disp_map.height_at(path)

# st = ts.Stimulus(location=loc, trace=trace)
# s = ts.stim_indent_shape(ts.shape_hand_region(region="D2d_t", pins_per_mm=0.5), trace, pin_radius=0.5)
pins = ts.shape_hand_region(region="D2d_t", pins_per_mm=0.5)
s = ts.stim_texture_indextip(disp_map, path, pins)

# plot(s)
```

```python
import json

with open("np.json", "w") as f:
    s = np.array(s).tolist()
    json.dump(s, f)
```

```python
plot(region='D2d') * plot(s, spatial=True, bin=10)
```

### Response

```python
r = a.response(s)
print(r)
```
